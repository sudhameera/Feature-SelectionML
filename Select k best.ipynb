```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('CKD.csv')
```

```python
data.isna().sum()
```

```python
def selectkbest(indep,dep,n):
    
    test=SelectKBest(score_func=chi2,k=n)
    fit1=test.fit(indep,dep)
    selectk_feature=fit1.transform(indep)
    return selectk_feature

```

```python
def split_scaler(indep,dep):
    x_train,x_test,y_train,y_test=train_test_split(indep,dep,test_size=0.25,random_state=0)
    sc=StandardScaler()
    x_train=sc.fit_transform(x_train)
    x_test=sc.transform(x_test)
    return x_train,x_test,y_train,y_test
```

```python
def cm_prediction(classifier,x_test):
    y_pred=classifier.predict(x_test)

    from sklearn.metrics import confusion_matrix
    cm=confusion_matrix(y_test,y_pred)
    from sklearn.metrics import accuracy_score
    from sklearn.metrics import classification_report
    accuracy=accuracy_score(y_test,y_pred)
    report= classification_report(y_test,y_pred)
    return cm,accuracy,report,x_test,y_test,classifier
```

```python
def logistic(x_train,y_train,x_test):
    from sklearn.linear_model import LogisticRegression
    classifier =LogisticRegression()
    classifier.fit(x_train,y_train)
    cm,accuracy,report,x_test,y_test,classifier=cm_prediction(classifier,x_test)
    return cm,accuracy,report,x_test,y_test,classifier
```

```python
def svm_linear(x_train,y_train,x_test):
    from sklearn.svm import SVC
    classifier =SVC(kernel='linear',random_state=0)
    classifier.fit(x_train,y_train)
    cm,accuracy,report,x_test,y_test,classifier=cm_prediction(classifier,x_test)
    return cm,accuracy,report,x_test,y_test,classifier
```

```python
def svmnl(x_train,y_train,x_test):
    from sklearn.svm import SVC 
    classifier = SVC(kernel='rbf',random_state=0)
    classifier.fit(x_train,y_train)
    cm,accuracy,report,x_test,y_test,classifier=cm_prediction(classifier,x_test)
    return cm,accuracy,report,x_test,y_test,classifier
```

```python
def Naive(x_train,y_train,x_test):
    from sklearn.naive_bayes import GaussianNB
    classifier=GaussianNB()
    classifier.fit(x_train,y_train)
    cm,accuracy,report,x_test,y_test,classifier=cm_prediction(classifier,x_test)
    return cm,accuracy,report,x_test,y_test,classifier
    
```

```python
def knn(x_train,y_train,x_test):
    from sklearn.neighbors import KNeighborsClassifier
    classifier = KNeighborsClassifier(n_neighbors=5,metric='minkowski' ,p=2)
    classifier.fit(x_train, y_train)
    cm,accuracy,report,x_test,y_test,classifier=cm_prediction(classifier,x_test)
    return cm,accuracy,report,x_test,y_test,classifier
    
```

```python
def des(x_train,y_train,x_test):
    from sklearn.tree import DecisionTreeClassifier
    classifier = DecisionTreeClassifier(criterion='entropy')
    classifier .fit(x_train,y_train)
    cm,accuracy,report,x_test,y_test,classifier=cm_prediction(classifier,x_test)
    return cm,accuracy,report,x_test,y_test,classifier
    
```

```python
def rf(x_train,y_train,x_test):
    from sklearn.ensemble import RandomForestClassifier
    classifier =RandomForestClassifier(n_estimators=10,criterion='entropy')
    classifier.fit(x_train,y_train)
    cm,accuracy,report,x_test,y_test,classifier=cm_prediction(classifier,x_test)
    return cm,accuracy,report,x_test,y_test,classifier
```

```python
def selectk_classification(acclog,accsvml,accsvmnl,accnaive,accknn,accdecision,accrandom):
    dataframe=pd.DataFrame(index=['chisquare'],columns=['Logistic','SVML','SVMNL','Naive','KNN','Decision','Random'])
    for number,index in enumerate(dataframe.index):
        dataframe['Logistic'][index]=acclog[number]
        dataframe['SVML'][index]=accsvml[number]
        dataframe['SVMNL'][index]=accsvmnl[number]
        dataframe['Naive'][index]=accnaive[number]
        dataframe['KNN'][index]=accknn[number]
        dataframe['Decision'][index]=accdecision[number]
        dataframe['Random'][index]=accrandom[number]
        return dataframe
```

```python
df1=data

df1=pd.get_dummies(df1,drop_first=True)
df1
```

```python
indep=df1.drop('classification_yes',axis=1)
dep=df1['classification_yes']
```

```python
kbest= selectkbest(indep,dep,5)
```

```python
acclog=[]
accsvml=[]
accsvmnl=[]
accnaive=[]
accknn=[]
accdecision=[]
accrandom=[]
```

```python
x_train,x_test,y_train,y_test=split_scaler(kbest,dep)
```

```python
cm,accuracy,report,x_test,y_test,classifier=logistic(x_train,y_train,x_test)
acclog.append(accuracy)

cm,accuracy,report,x_test,y_test,classifier=svm_linear(x_train,y_train,x_test)
accsvml.append(accuracy)

cm,accuracy,report,x_test,y_test,classifier=svmnl(x_train,y_train,x_test)
accsvmnl.append(accuracy)

cm,accuracy,report,x_test,y_test,classifier=Naive(x_train,y_train,x_test)
accnaive.append(accuracy)

cm,accuracy,report,x_test,y_test,classifier=knn(x_train,y_train,x_test)
accknn.append(accuracy)

cm,accuracy,report,x_test,y_test,classifier=des(x_train,y_train,x_test)
accdecision.append(accuracy)

cm,accuracy,report,x_test,y_test,classifier=rf(x_train,y_train,x_test)
accrandom.append(accuracy)
```

```python
result=selectk_classification(acclog,accsvml,accsvmnl,accnaive,accknn,accdecision,accrandom)
result
#3
```
